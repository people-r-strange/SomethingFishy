### DOB has to be greater than voter registration by at least 18 years, is this true? 

No this is not true, there are 84 voters whose birth dates did not align with when they were suppose to register to vote. 

```python
#changing date format to match the rest of the dates
df["DOB"].replace({"8-Apr-34": "1934-04-08"},inplace=True)
```

```python
#changing the DOB dates to datetime type
df['DOB_2'] =  pd.to_datetime(df['DOB'])
#https://stackoverflow.com/questions/29310116/removing-time-from-datetime-variable-in-pandas
```

```python
#converting the registration dates to datetime type
df['VOTER REG DATE'] =  pd.to_datetime(df['VOTER REG DATE'])
```

```python
#finding the difference between registration date and dob
df['age_in_days'] = (df['VOTER REG DATE'] - df['DOB_2'])
```

```python
#converting the difference from days to years (approximately)
df['age_in_yrs'] = df['age_in_days'] / 365
```
